+ 생성자(Constructor) 메서드와 소멸자(Deconstructor) 메서드(함수)
 - 객체를 생성하는 용도 
 - 객체를 처음 만들 때 호출된다. (단 한번만 호출됨)
 - 초기화가 가능하다.
 

```python
class Test:
    a = int()
    b = str()
    
    def __init__(self,x=0,y=None): # 생성자 함수
        print("생성자 호출됨")
        self.a = x
        self.b = y
        
    def setValue(self,x,y):
        self.a = x
        self.b = y

    
    def __del__(self): # 소멸자 함수
        print("소멸자 호출됨")

class Test2:
    pass

print("-"*35)
t = Test(10,"태양")
print(t.a,t.b)

t.setValue(20,"지디")
print(t.a,t.b)

t1 = Test()
print(t1.a,t1.b)

t2 = Test(30)
print(t2.a,t2.b)

t3 = Test("탑")
print(t3.a,t3.b)

del t3

print("-"*35)
ob1 = Test()
ob2 = Test2()

print(type(ob1),type(ob2))

print(isinstance(ob1, Test))
print(isinstance(ob1, Test2))
print(isinstance(ob2, Test))
print(isinstance(ob2, Test2))
```

+ 정적 메서드 <-> 인스턴스 메서드
 - 정적(클래스) 변수 / 정적(클래스) 메서드
 - 동일한 공간을 모든 객체들이 공동으로 사용 
 - staticMethod, classMethod
 
    

```python
class Person:
    nation = "한국"
    name = str()
    age = int()
    height = float()
    hobby = []
    telphone = []
    
    def show(self):
        print("국적 : ",self.nation)
        print("이름 : ",self.name)
        print("나이 : ",self.age)
        print("신장 : ",self.height)
        print("취미 : ",self.hobby)
        print("전화번호 : ",self.telphone)
        print("=================================")
    
    @staticmethod
    def tel(num="010-1111-1111"):
        Person.telphone.append(num)
        return "대표 전화번호 : " + Person.telphone[0]

#######################################################

print(Person.nation)

p1 = Person()
p2 = Person()
print(p1.nation, p2.nation)

p1.nation = "독일"
p2.nation = "프랑스"
print(p1.nation, p2.nation)
print(Person.nation)


print(Person.tel())
print(p1.tel())
print(p2.tel())
```

```python
# 객체의 수를 카운팅

class Count:
    cnt = 0
    
    def __init__(self):
        Count.cnt += 1
        
    @staticmethod
    def show():
        print("객체의 수 : ", Count.cnt)
        
####################################

a,b,c = Count(),Count(),Count()
a.show()
b.show()
c.show()

Count.show()
```

```python
class Count:
    cnt = 0
    
    def __init__(self):
        Count.cnt += 1
        
    @staticmethod
    def show():
        print("객체의 수 : ", Count.cnt)
    
    @classmethod
    def show2(cls): # ()클래스 본체의 주소
        print("객체의 수 : ", cls.cnt)
        
####################################

a,b,c = Count(),Count(),Count()
a.show2()
b.show2()
c.show2()

Count.show2()
```

```python
# decorator를 사용하지 않을 경우

class Count:
    cnt = 0
    
    def __init__(self):
        Count.cnt += 1
        
    def show():
        print("객체의 수 : ", Count.cnt)
    
    def show2(cls): # ()클래스 본체의 주소
        print("객체의 수 : ", cls.cnt)

    
    cnt1 = staticmethod(show)
    cnt2 = classmethod(show2)
    
####################################

a,b,c = Count(),Count(),Count()
a.cnt1()
b.cnt1()
c.cnt1()

a.cnt2()
b.cnt2()
c.cnt2()

Count.cnt1()
Count.cnt2()
```

### 상속(Inheritance)
+ 상속 설계시 고려할 점
    - 부모 클래스 : Super class / Base class(기본)
    - 자식 클래스 : Sub class / Derived class(파생)
    - 자식 클래스가 공통적으로 필요로 하는 것들로 구성.
    - 부모 클래스는 최소한의 코드로 작성.
    - 상속 받은 자식 클래스는 반드시 기능을 확장(추가) 시켜야한다.
    - 중복된 기능이 있어서는 안된다.
    
    
+ ⭐️Override(재정의) , Overload(중복정의)
    - override는 반드시 상속에서만 사용.
    - overridesms 부모 클래스에서 정의한 메서드를 자식 클래스에서 변경하는 것.
    - overload는 상속과 상관없이 어디에서든 사용.
    - overload는 파이썬에서 지원하지 않는다.

```python
class Animal:
    def move(self):
        print("움직인다.")
        
class Dog(Animal):
    def bart(self):
        print("멍멍!!!!")

    
class Horse(Animal):
    def speed(self,a):
        print("현재 속도는",a)
        
d1 = Dog()
d1.move()
d1.bart()

h1 = Horse()
h1.move()
h1.speed(1000)
```

### Override

```python
class Calculator:
    first = 0
    second = 0
    
    def __init__(self,first=None,second=None):
        self.first = first
        self.second = second
        
    def setData(self,first=None,second=None):
        self.first = first
        self.second = second
    
    def sum(self):
        return self.first + self.second
    
    def sub(self):
        return self.first - self.second
    
    def mul(self):
        return self.first * self.second
    
    def div(self):
        return self.first / self.second
    

class Calculator2(Calculator):
    def pow(self):
        return self.first ** self.second
    
    def div(self):
        if self.first == 0 or self.second == 0:
            return "숫자가 아닙니다."
        else:
            return self.first / self.second

    
cal1 = Calculator2()
cal1.setData(10,20)
print(cal1.sum())

cal2 = Calculator()
cal2.setData(20,30)
print(cal2.mul())

cal3 = Calculator2()
cal3.setData(2,10)
print(cal3.pow())

cal3.setData(10,0)
print(cal3.div())
```

### super 
   + 부모의 주소를 받아오는 메서드 : (super().~)

```python
class Person:
    say = "난 사람이다."
    age = 20
    
    def info(self):
        print("나이:{}. 하고싶은 말:{}".format(self.age,self.say))
        
    def __init__(self):
        print("Person 클래스")
    
class Employee(Person):
    say = "난 회사원이다."
    
    def __init__(self):
        print("Employee 클래스")
        
    def info(self):
        print(self.say)
        print(self.age)
        print(super().say)
        super().info()
        
################################################

e1 = Employee()
e1.info()
```

### 추상 클래스와 추상 메서드
+ 추상 클래스 : 추상 메서드를 한 개라도 가지고 있으면 추상 클래스가 된다.
    - 인스턴스를 만들수 없음(일을 안하겠다는 뜻)
    - metaclass=abc.ABCMeta : 부모클래스가 메타클래스를 반드시 상속을 받아야함

    

+ 추상 메서드 : 메서드안에 내용이 비워져 있어야 한다.
    - @abc.abstractmethod : 반드시 오버라이딩을 제대로 해야한다.

```python
import abc

class Volume(metaclass=abc.ABCMeta):
    vol = 0
    
    @abc.abstractmethod
    def volumeUp(self,vol):
        pass
    
    @abc.abstractmethod
    def volumeDown(self,vol):
        pass
    
class TestVolume(Volume):
    def volumeUp(self,vol):
        self.vol += vol
        print("테스트 제품 소리 증가 :",self.vol)
    
    def volumeDown(self,vol):
        self.vol -= vol
        print("테스트 제품 소리 감소 :",self.vol)
        
###############################################

t = TestVolume()
t.volumeUp(100)
t.volumeDown(50)
```

```python
"""
1. 부모 클래스로 Power를 추가한다. : 다중 상속 구현
    - powerOn(), powerOff()

2. 부모 클래스는 추상 클래스로 구현한다.

3. 전원을 켰을 때만 볼륨 조정을 할 수 있게 한다.

4. 볼륨 조정은 max값을 100으로 하고 min값을 0으로 한다.

-----------------------------------------------------------
1. TV
2. Radio
3. Speaker
4. 종료
메뉴 : 1

1. 전원
2. 볼륨 조정
3. 나가기
메뉴 : 2

볼륨 레벨 : 20
현재 TV의 볼륨은 20입니다.
...

"""
import abc

class Power(metaclass=abc.ABCMeta):
    switch = False

    @abc.abstractmethod
    def powerOn(self):
        self.switch = True
        print("전원을 켰습니다.")

    @abc.abstractmethod
    def powerOff(self):
        self.switch = False
        print("전원을 껐습니다.")


##############################################################################


class Volume(metaclass=abc.ABCMeta):
    vol = 0
    mx = 100
    mn = 0

    @abc.abstractmethod
    def volumeUp(self, vol):
        self.vol += vol
        print("볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다." % (vol, self.vol))

    @abc.abstractmethod
    def volumeDown(self, vol):
        self.vol -= vol
        print("볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다." % (vol, self.vol))


############################################################################

class TV(Volume, Power):
    def powerOn(self):
        self.switch = True
        print("TV의 전원을 켰습니다.")

    def powerOff(self):
        self.switch = False
        print("TV의 전원을 껐습니다.")

    def volumeUp(self, vol):
        if self.switch == True:
            if 0 <= self.vol+vol <= 100:
                self.vol += vol
                print("TV 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다." % (vol, self.vol))
            elif self.vol+vol > 100:
                print("TV의 볼륨을 높일 수 있는 최대값을 넘었습니다. 다시 입력해주세요.")

        elif self.switch == False:
            print("TV의 전원을 켜주세요")

    def volumeDown(self, vol):
        if self.switch == True:

            if 0 <= self.vol-vol <= 100:
                self.vol -= vol
                print("TV 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다." % (vol, self.vol))
            elif self.vol-vol < 0:
                print("TV의 볼륨을 줄일 수 있는 최소값을 넘었습니다. 다시 입력해주세요.")

        elif self.switch == False:
            print("TV의 전원을 켜주세요")

##############################################################################

class Radio(Volume, Power):
    def powerOn(self):
        self.switch = True
        print("Radio의 전원을 켰습니다.")

    def powerOff(self):
        self.switch = False
        print("Radio의 전원을 껐습니다.")

    def volumeUp(self, vol):
        if self.switch == True:
            if 0 <= self.vol + vol <= 100:
                self.vol += vol
                print("Radio 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다." % (vol, self.vol))
            elif self.vol + vol > 100:
                print("Radio의 볼륨을 높일 수 있는 최대값을 넘었습니다. 다시 입력해주세요.")

        elif self.switch == False:
            print("Radio의 전원을 켜주세요")

    def volumeDown(self, vol):
        if self.switch == True:

            if 0 <= self.vol - vol <= 100:
                self.vol -= vol
                print("Radio 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다." % (vol, self.vol))
            elif self.vol - vol < 0:
                print("Radio의 볼륨을 줄일 수 있는 최소값을 넘었습니다. 다시 입력해주세요.")

        elif self.switch == False:
            print("Radio의 전원을 켜주세요")

##############################################################################

class Speaker(Volume, Power):

    def powerOn(self):
        self.switch = True
        print("Speaker의 전원을 켰습니다.")

    def powerOff(self):
        self.switch = False
        print("Speaker의 전원을 껐습니다.")

    def volumeUp(self, vol):
        if self.switch == True:
            if 0 <= self.vol + vol <= 100:
                self.vol += vol
                print("Speaker 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다." % (vol, self.vol))
            elif self.vol + vol > 100:
                print("Speaker의 볼륨을 높일 수 있는 최대값을 넘었습니다. 다시 입력해주세요.")

        elif self.switch == False:
            print("Speaker의 전원을 켜주세요")

    def volumeDown(self, vol):
        if self.switch == True:

            if 0 <= self.vol - vol <= 100:
                self.vol -= vol
                print("Speaker 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다." % (vol, self.vol))
            elif self.vol - vol < 0:
                print("Speaker의 볼륨을 줄일 수 있는 최소값을 넘었습니다. 다시 입력해주세요.")

        elif self.switch == False:
            print("Speaker의 전원을 켜주세요")

class TV_view(TV):
    def tv_view(self):
        while True:
            print("-----------")
            print("1. 전원")
            print("2. 볼륨")
            print("3. 나가기")
            print("-----------")
            self.menu = int(input("메뉴 : "))

            if self.menu == 1:
                print("-----------")
                print("1. 전원 켜기")
                print("2. 전원 끄기")
                print("3. 이전")
                print("-----------")
                self.menu = int(input("메뉴 : "))

                if self.menu == 1:
                    print("-----------")
                    self.TV.powerOn()
                elif self.menu == 2:
                    print("-----------")
                    self.TV.powerOff()
                elif self.menu == 3:
                    continue
                else:
                    print("-----------")
                    print("잘못 입력하였습니다.다시 입력해주세요.")


            elif self.menu == 2:
                print("-----------")
                print("1. 볼륨 올리기")
                print("2. 볼륨 내리기")
                print("3. 이전")
                print("-----------")
                self.menu = int(input("메뉴 : "))

                if self.menu == 1:
                    print("-----------")
                    self.TV.volumeUp(int(input("볼륨 증가량 : ")))

                elif self.menu == 2:
                    print("-----------")
                    self.TV.volumeDown(int(input("볼륨 감소량 : ")))

                elif self.menu == 3:
                    continue
                else:
                    print("-----------")
                    print("잘못 입력하였습니다.다시 입력해주세요.")

            elif self.menu == 3:
                break

            else:
                print("-----------")
                print("잘못 입력하였습니다.다시 입력해주세요.")


class Radio_view(Radio):
    def radio_view(self):
        while True:
            print("-----------")
            print("1. 전원")
            print("2. 볼륨")
            print("3. 나가기")
            print("-----------")
            self.menu = int(input("메뉴 : "))

            if self.menu == 1:
                print("-----------")
                print("1. 전원 켜기")
                print("2. 전원 끄기")
                print("3. 이전")
                print("-----------")
                self.menu = int(input("메뉴 : "))

                if self.menu == 1:
                    print("-----------")
                    self.Radio.powerOn()
                elif self.menu == 2:
                    print("-----------")
                    self.Radio.powerOff()
                elif self.menu == 3:
                    break
                else:
                    print("-----------")
                    print("잘못 입력하였습니다.")

            elif self.menu == 2:
                print("-----------")
                print("1. 볼륨 올리기")
                print("2. 볼륨 내리기")
                print("3. 이전")
                print("-----------")
                self.menu = int(input("메뉴 : "))

                if self.menu == 1:
                    print("-----------")
                    self.Radio.volumeUp(int(input("볼륨 증가량 : ")))

                elif self.menu == 2:
                    print("-----------")
                    self.Radio.volumeDown(int(input("볼륨 감소량 : ")))

                elif self.menu == 3:
                    break
                else:
                    print("-----------")
                    print("잘못 입력하였습니다.다시 입력해주세요.")

            elif self.menu == 3:
                break
            else:
                print("-----------")
                print("잘못 입력하였습니다.다시 입력해주세요.")

class Speaker_view(Speaker):
    def speaker_view(self):
        while True:
            print("-----------")
            print("1. 전원")
            print("2. 볼륨")
            print("3. 이전")
            print("-----------")
            self.menu = int(input("메뉴 : "))

            if self.menu == 1:
                print("-----------")
                print("1. 전원 켜기")
                print("2. 전원 끄기")
                print("3. 이전")
                print("-----------")
                self.menu = int(input("메뉴 : "))

                if self.menu == 1:
                    self.Speaker.powerOn()
                elif self.menu == 2:
                    self.Speaker.powerOff()
                elif self.menu == 3:
                    break
                else:
                    print("-----------")
                    print("잘못 입력하였습니다.다시 입력해주세요.")

            elif self.menu == 2:
                print("-----------")
                print("1. 볼륨 올리기")
                print("2. 볼륨 내리기")
                print("3. 이전")
                print("-----------")
                self.menu = int(input("메뉴 : "))

                if self.menu == 1:
                    print("-----------")
                    self.Speaker.volumeUp(int(input("볼륨 증가량 : ")))
                elif self.menu == 2:
                    print("-----------")
                    self.Speaker.volumeDown(int(input("볼륨 감소량 : ")))
                elif self.menu == 3:
                    break
                else:
                    print("잘못 입력하였습니다.다시 입력해주세요.")
            elif self.menu == 3:
                break

            else:
                print("-----------")
                print("잘못 입력하였습니다.다시 입력해주세요.")

##############################################################################

class View(TV_view, Radio_view, Speaker_view):
    def __init__(self):
        self.TV = TV()
        self.Radio = Radio()
        self.Speaker = Speaker()

    def display(self):
        while True:
            print("-----------")
            print("1. TV")
            print("2. Radio")
            print("3. Speaker")
            print("4. 종료")
            print("-----------")
            self.menu = int(input("메뉴 : "))

            if self.menu == 1:
                self.tv_view()

            elif self.menu == 2:
                self.radio_view()

            elif self.menu == 3:
                self.speaker_view()

            elif self.menu == 4:
                print("종료되었습니다.")
                break
            else:
                print("-----------")
                print("잘못 입력하였습니다. 다시 입력해주세요.\n")

##############################################################################

view = View()
view.display()
```

### 클래스의 관계

+ is a 관계(상속 관계 , 일반화 관계)
    - 특정 클래스 객체가 다른 클래스 객체에게 자신의 멤버를 물려주는 상속관계
    - UML에서는 비어있는 화살표로 강한 결합 관계로 표현

+ has a 관계(연관 관계, 포함 관계)
    - 특정 클래스내에서 해당 클래스의 멤버처럼 다른 클래스 객체를 포함하고 있는 경우
    - Aggregation : 약한 관계
    - Composition : 강한 관계
    - UML에서는 마름모로 표현
    
+ UML

```python
# has a 관계 중 강한 관계

class Handle:
    quantity = 0
    
    def leftTurn(self,q):
        self.quantity = q
        return "좌회전"
    
    def rightTurn(self,q):
        self.quantity = q
        return "우회전"
    
    def straight(self,q):
        self.quantity = q
        return "직진"
    
#####################################################
    
class Car:
    turnShow = "정지"
    
    def __init__(self,owner):
        self.owner = owner
        self.handle = Handle()
    
    def turnHandle(self,q):
        if q > 0:
            self.turnShow = self.handle.rightTurn(q)
        elif q < 0:
            self.turnShow = self.handle.leftTurn(q)
        elif q == 0:
            self.turnShow = self.handle.straight(q)
            
        
#####################################################

tom = Car("tom")
tom.turnHandle(10)
print(tom.owner + "의 회전량은 "+ tom.turnShow,str(tom.handle.quantity))
```

```python
# has a 관계 중 약한 관계

class Fridge:
    isOpen = False
    foods = []
    
    def open(self):
        self.isOpen = True
        print("냉장고 문이 열렸습니다.")
    
    def close(self):
        self.isOpen = False
        print("냉장고 문이 닫혔습니다.")
    
    def put(self,thing):
        if self.isOpen:
            self.foods.append(thing)
            print("냉장고 음식이 들어감")
            self.list()
        else:
            print("냉장고 문이 닫혀있어서 음식을 넣을 수 없음")
    
    def list(self):
        for f in self.foods:
            print("-", f.name, f.ex_data)
        print()
###################################

class FoodData:
    def __init__(self, name, ex_data):
        self.name = name
        self.ex_data = ex_data
        
###################################

f = Fridge()
apple = FoodData("사과","2021-5-5")
bro = FoodData("브로콜리", "2021-5-1")

f.open()
f.put(apple)
f.put(bro)
f.close()
        
```

#### (2) 클로저(Closure)

+ 함수 밖에서 임의의 함수 안에 선언된 지역변수를 계속 참조할 수 있게 하는 방법


```python
count = 0

def out():
    count = 0
    
    def inn():
        global count
        #nonlocal count
        count += 1 
        print(count)

    inn()
    
######################################

out()
out()
out()
```

```python
def out():
    count = 0
    
    def inn():
        nonlocal count
        count += 1 
        print(count)

    return inn
    
######################################

a = out()
a()
a()
a()
print(id(a))

b = out()
b()
b()
b()
print(id(b))
```

```python
def outer(tax):
    def inner(su,dan):
        amount = su * dan * tax
        return amount
    
    return inner
    
################################

tax_rate = outer(0.1)
mouse = tax_rate(5,5000)
print(mouse)

usb = tax_rate(3, 12000)
print(usb)

tax_rate2 = outer(0.05)
mouse = tax_rate2(5, 5000)
print(mouse)
```

#### (4) 함수 장식자(decorator)

+ 일종의 함수 감싸기(wrapping)을 해주는 디자인 패턴


+ 메타 프로그래밍 기법


+ @classmethod, @staticmethod, @abstract, ...

```python
def make2(fa):
    f = lambda:"안녕 " + fa()
    return f

def make1(fa):
    f = lambda:"반가워 " + fa()
    return f


@make2
@make1
def hello():
    return "지디"

#############################

hi = hello
print(hi())
```

```python
def outer(fn):
    def inner(n1,n2):
        print("결과는 {}이다.".format(fn(n1,n2)))
    return inner


@outer
def func(n1,n2):
    return n1 + n2

#############################################

# result = func
# result(10, 20)

func(10,20)
```
