+ SQL(Structured Query Language)
    * DDL(Data Definition Language) : 데이터 정의어
       - CREATE, ALTER, DROP
       
    * DML(Data Manipulation Language) : 데이터 조작어
       - (SELECT), INSERT, UPDATE, DELETE
       
    * QL(Query Language) : 데이터 질의어
       - SELECT
       
    * DCL(Data Controll Language) : 데이터 제어어(보안, 인증, 시스템관리, 트랜젝션 관리 등)
       - GRANT, REVOKE, ...
       
    * 표준 SQL(ANSI SQL)
    
    * 제품에 따른 SQL
       - Oracle : PL/SQL
       - MS : T-SQL

+ SELECT
   * 기본 문법
     - SELECT 필드명1, 필드명2 FROM 테이블명;
     - SELECT * FROM 테이블명;
        - SELECT ename, sal, deptno, job, FROM emp;
        - SELECT * FROM emp;
     
     - 필드의 가공 처리 
        - SELECT empno, ename,sal,sal+200 FROM emp;
     
     - 필드의 alias
        - SELECT ename, mgr as manager, sal+200 as "add salary" FROM emp;
        - SELECT ename, mgr manager, sal+200 "add salary" FROM emp;
     
     - 중복 제거
        - SELECT distinct job FROM emp;
        
   * 확장 문법
      - 정렬 기능 : ORDER BY 필드명 [ASC(오름차순) | DESC(내림차순)], 필드명,...(무조건 맨 마지막에 작성)
        - 급여가 많은 순으로 이름, 급여, 부서코드 조회
            - SELECT ename,sal,deptno FROM emp ORDER BY sal DESC;
        
        - 부서별로 급여가 적은 순으로 이름, 급여, 부서코드 조회
            - SELECT ename,sal,deptno FROM emp ORDER BY sal,deptno;
      
      - 조건 기능(레코드 필터링) : WHERE 조건문
        
        - 급여가 3000원 이상인 직원의 사번, 이름, 업무, 급여를 조회
            - SELECT ename, deptno, job, sal FROM emp WHERE = sal >= 3000;
        
        - 업무가 manager인 직원의 이름, 부서, 업무, 급여를 조히
            - SELECT ename, deptno, job, sal FROM emp WHERE job = 'manager'

        - 부서가 20이고 업무가 analyst인 직원의 이름, 부서, 업무, 급여를 조회
            - SELECT ename, deptno, job, sal FROM emp WHERE deptno = 20 AND job = 'analyst';

        - 급여가 1500이상 2500 이하를 받는 직원의 이름, 부서, 업무, 급여를 조회
            - SELECT ename, deptno, job, sal FROM emp WHERE sal >= 1500 and sal <= 2500;
            - SELECT ename, deptno, job, sal FROM emp WHERE sal BETWEEN 1500 and 2500;
        
        - 급여가 1500이상 2500이하를 받는 직원의 이름, 부서, 업무, 급여를 조회 단, 급여가 많은 순으로 조회
            - SELECT ename, deptno, job, sal FROM emp WHERE sal BETWEEN 1500 and 2500 ORDER BY sal DESC;
        
        - 업무가 clerk, salesman, analyst인 직원의 이름, 부서, 업무, 급여를 조회
            - SELECT ename, deptno, job, sal FROM emp WHERE job = 'clerk' or job = 'salesman' or job = 'analyst';
            - SELECT ename, deptno, job, sal FROM emp WHERE job IN('clerk','salesman','analysy');
        
        - 1982년 1월 1일 이후에 입사한 직원의 이름, 업무, 입사일자를 조회
            - SELECT ename, job, hiredate FROM emp WHERE hiredate >= '1982-1-1';
        
        - 커미션이 없는 직원의 이름, 부서, 업무, 급여, 커미션을 조회
            - SELECT ename, deptno, job, sal, comm FROM emp WHERE comm NOT null;
            - SELECT ename, deptno, job, sal, comm FROM emp WHERE job != 'salesman';
        
        - 커미션이 있는 직원의 이름, 부서, 업무, 급여, 커미션을 조회
            - SELECT ename, deptno, job, sal, comm FROM emp WHERE comm IS NOT null;
        
     - LIKE 연산자
        - % : 0개 이상의 문자를 대체
        - _ : 1개 문자를 대체하는 기호
        
        - 이룸이 s로 시작하는 직원의 이름, 업무, 급여를 조회
           - SELECT ename, job, sal FROM emp WHERE ename LIKE 's%';
        
     - GROUP BY 필드명 [,필드명.필드명,...][HAVING 조건문]
        - 부서별로 급여 평균, 최저 급여를 조회
           - SELECT deptno,avg(sal),min(sal) FROM emp GROUP BY deptno;
        
        - 부서별로 급여평균, 최저 급여를 조회(단, 급여평균이 높은 순으로 조회)
           - SELECT deptno,avg(sal) as '급여평균',min(sal) as '최저급여'FROM emp GROUP BY deptno ORDER BY avg(sal)DESC;
        
        - 전체 급여의 합계가 5000을 초과하는 업무에 대해 급여 합계를 조회
           - SELECT job, sum(sal) as '전체 급여' FROM emp GROUP BY job HAVING sum(sal) > 5000;
        
        - 전체 급여의 합계가 5000을 초과하는 업무에 대해 급여 합계를 조회 단, salesman은 제외
           - SELECT job, sum(sal) as '전체 급여' FROM emp GROUP BY job HAVING sum(sal) > 5000 and job != 'salesman';
           - SELECT job, sum(sal) as '전체 급여' FROM emp WHERE job != 'salesman' GROUP BY job HAVING sum(sal) > 5000;

+ 함수(Single Row Function, Aggregation Function, ....)
    - Single Row Function(String, Numeric, Data&Time)
      - String Function
         - ASCII(), CHAR()
           - SELECT ascii('a'),char(98);
        
         - LENGTH()
           - SELECT ename, length(ename) FROM emp;
        
         - INSTR()
            - SELECT instr('foobarbar','bar');
        
         - LEFT(), RIGHT(), SUBSTR(), SUBSTRING()
            - SELECT left('abcdefg',3);       
            - SELECT substring('l love sql',3);
            - SELECT substring('l love sql',3,4);
            - SELECT substring('l love sql',-1);
            - SELECT substring('l love sql',-8,-3);
            
        - LOWER(), UPPER()
            - SEKECT lower('ABcdEF'),UPPER('ABcdEF');
        
        - LTRIM(), RTRIM(), TRIM()
            - SELECT '         abc          ';
            - SELECT LTRIM('         abc          ') as trim;
            - SELECT RTRIM('         abc          ') as  trim;
            - SELECT TRIM('         abc          ') as trim;
            
    - Numeric Function
        - CEIL(), FLOOR(), ROUND(), TRUNCATE()
            - SELECT ceil(1.23), floor(1.78);
            - SELECT round(1.23), round(1.23,0), round(1.78,1);
            - SELECT truncate(1.78,0), truncate(1.78,1), truncate(1.78, -1); 
            
        - MOD()
            - SELECT mod(29,0),20 mod 9,20 % 9;
            
        - RAND()
            - SELECT rand();
            
            /* n <= R m
               floor(rand() * (m-n) + n)*/
               
            - 7보다 크거나 같고 12보다 작은 범위내에서 난수를 뽑아라.
            - SELECT floor(rand() * (12-7) + 7);
             
    - Date & Time Function
        - NOW(), SYSDATE(), CURDATE(), CURTIME()
        
           - YEAR(), MONTH(), DAOFMONTH() : 연, 월, 일
             - SELECT year(now()),month(now()), dayofmonth(now());
             - SELECT dayofmonth('2021/05/19');
        
           - DAYOFWEEK() : 1-일요일 , WEEKDAY() : 0-월
             - SELECT dayofweek(now()), weekday(now());
            
           - DAYOFYEAR() : 현재까지의 일수
             - SELECT dayofyear(now());
            
           - DATE_ADD(), DATE_SUB(), ADDDATE()
             - SELECT DATE_ADD(now(), interval 5 year);
             - SELECT DATE_ADD(now(), interval 100 day);
             
           - 그 외
             - SELECT datediff(now(), '2020/12/31');
             - SELECT last_day(now());

    - Aggregation Function : avg(), sum(), count(), max(), min()

        - WHERE절에는 그룹함수를 사용할 수 없다.

            - SELECT count(comm) FROM emp;
            
    - 기타 함수
        - CONVERT( )

        - CASE

+ Sub Query 
    - 다른 Query문에 포함된 Query문
        - scott의 급여보다 더 많이 받는 직원의 이름, 업무, 급여를 조회
            - SELECT ename, job, sal FROM emp WHERE ename = 'scott';
            - SELECT ename, job, sal FROM emp WHERE sal>(select sal from emp where ename = 'scott');

        - 사번이 7521의 업무와 같고, 급여가 7943보다 많은 직원의 사번, 이름, 업무, 급여를 조회
            - SELECT empno, ename, job, sal FROM emp WHERE job=(select job from emp where empno = 7521) and sal>(select sal from emp where empno=7934);
        
        - 가장 최근에 입사한 직원의 이름, 부서, 업무, 입사일자 조회
            - SELECT ename, deptno, job, hiredate FROM emp WHERE hiredate = "870523";
            - SELECT ename, deptno, job, hiredate FROM emp WHERE hiredate=(SELECT max(hiredate) FROM emp);

    - 반드시 ( ) 사용
   
    - 연산자의 오른쪽에 와야 한다.
   
    - 내부 Query는 ORDER BY 사용 불가
   
    - 연산자
        - 단일행 연산자 : 기본 연산자
        - 다중행 연산자 : IN(=), ANY(or), ALL(and), EXISTS ...
            - 업무별 최소 급여를 받는 직원의 사번, 이름, 부서코드를 조회

                - SELECT empno, ename, deptno, min(sal) FROM emp WHERE sal IN(800,1250,2450,3000,5000);
                - SELECT empno, ename, deptno FROM emp WHERE sal IN(SELECT min(sal) FROM emp GROUP BY job);
                
            - 업무별 최소급여보다 많이 받는 직원의 사번, 이름, 부서코드를 조회
                - SELECT empno, ename, deptno max(sal) FROM emp WHERE sal > ANY(SELECT min(sal) FROM emp GROUP by job);
            - 업무별 최대급여 이상 받는 직원의 사번, 이름, 부서코드를 조회
                - SELECT empno, ename, deptno FROM emp WHERE sal >= ALL(SELECT max(sal) FROM emp GROUP BY job);
                
     - 다중열 비교 
         - 다중으로 서브쿼리를 실행
            - 'miller'의 급여와 보너스를 1500,300로 변경
                - UPDATE emp SET sal=1500, comm=300 where ename='miller';
                
            - 급여와 커미션이 30번 부서에 있는 직원들의 급여와 커미션이 같은 다른 부서 직원에 대해 사번,이름,부서번호,급여,커미션을 조회

                - SELECT empno,ename,deptno,sal,comm FROM emp WHERE (sal,comm)IN (SELECT sal,comm FROM emp WHERE deptno=30);
                
     - 상관 Sub Query

+ JOIN
    - 여러 개의 테이블을 병합하여 하나의 결과를 도출하기 위한 방법
    - 종류
        - Cartesian Product Join(데카르트 곱 조인)
        
        - Equi Join
            - 공통 필드의 레코드를 가져오는 방법(중복)
            - INNER JOIN(중복 제거)
            
        - INNER JOIN
            - SELECT 필드명 FROM 테이블1 INNER JOIN 테이블2 ON 조건문;
            - SELECT a.id,a.value FROM tblA a INNER JOIN tblB b ON a.id = b.id;
            - SELECT a.id,a.value FROM tblA a JOIN tblB b ON a.id = b.id;
            - SELECT a.id,a.value FROM tblA a,tblB b WHERE a.id = b.id;
            
            - 직원의 사번,이름,업무,부서번호,부서명을 조회
                - SELECT empno,ename,job,dname FROM emp JOIN dept ON emp.deptno = dept.deptno;
            
            - salesman의 업무를 수행하는 직원의 사번,이름,업무,부서번호,부서명을 조회
                - SELECT empno,ename,job,dname FROM emp JOIN dept ON emp.deptno = dept.deptno and job = 'salesman'
            
            - New york에서 근무하는 직원의 사번,이름,업무,부서명 조회
                - SELECT empno,ename,job,dname FROM emp JOIN dept ON emp.deptno = dept.deptno and loc = 'new york';
                
        - OUTER JOIN
            - INNER JOIN의 확장 : INNER JOIN + 공통되지 않은 레코드로 추출
            
            - 종류 : LEFT OUTER JOIN : 왼쪽에 있는건 전부 받음, RIGHT OUTER JOIN : 오른쪽에 있는건 전부 받음, FULL OUTER JOIN : 양쪽 모두 받음(MYSQL에서는 지원을 하지 않는다.)
            
            - SELECT 필드명1, 필드명2... FROM 테이블 LEFT OUTER JOIN 테이블2 ON 조건문;
            - SELECT a.id, a.value FROM tblA a LEFT OUTER JOIN tblB b ON a.id=b.id;
            - SELECT a.id, a.value, b.id, b.value FROM tblA a LEFT OUTER JOIN tblB b ON a.id = b.id;
            - SELECT a.id, a.value, b.id, b.value FROM tblA a RIGHT OUTER JOIN tblB b ON a.id = b.id;
            
            - 직원들의 이름,급여,부서명,근무지를 조회하시오. 단, 부서명과 근무지는 모두 출력할 수 있도록 하시오.
                - SELECT ename,sal,dname,loc FROM emp RIGHT OUTER JOIN dept ON emp.deptno = dept.deptno;
                
        - Non Equi Join
            - 서로 공통되지 않는 필드의 레코드를 가져오는 방법
            
            - 직원들의 사번,이름,업무,급여,급여 등급 조회
                - SELECT empno,ename,job,sal,grade FROM emp INNER JOIN salgrade ON emp.sal >= salgrade.losal and emp.sal <= salgrade.hisal;
                
        - Slef Join
            - 서로 같은 테이블을 복제하여 조인
            
            - 직원의 사번,이름,업무,관리자,관리자 이름을 조회
                - SELECT empno,ename,job,mgr,ename FROM emp a INNER JOIN emp b ON a.mgr = b.empno;
               
        - 세 개의 테이블 조인
            - 세 개의 테이블을 조인해서 a=b=c 처럼 동일한 값만 뽑는다.
                - SELECT a.id,b.value FROM tbla a INNER JOIN tblb b ON a.id = b.id INNER JOIN tblc c ON b.id = c.id;
